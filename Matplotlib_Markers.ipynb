fmt -> Format String


> Shortcut string notation way synatx:
marker | line | color



```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array ([2,3,4,5,67,8])

plt.plot(y,'o:r')
plt.show()
```

```python
plt.plot(y,'o-r')
plt.show()
```

```python
plt.plot(y,'o--r')
plt.show()
```

```python
plt.plot(y,'o-.r')
plt.show()
```

 If you leave out the line value in the fmt parameter, no line will be plotted.

Marker Size

> Its an keyword argument ms is its short version






```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([2,5,4,7,8])

plt.plot(y, marker= 'o', ms =20)
plt.show()
```

Marker Color


> keyword argument markeredgecolor or the shorter mec



```python
plt.plot(y,marker ="*",ms = 10, mec ='r')
```

keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers

```python
plt.plot(y ,marker='*',ms = 10,mfc = 'r')
plt.show()
```

```python
plt.plot(y ,marker='*',ms = 10,mec='b' ,mfc = 'r')
plt.show()
```

```python
plt.plot(y ,marker='*',ms = 10,mec='#4CAF50',mfc = 'r')
plt.show()
```

```python
plt.plot(y ,marker='*',ms = 10,mec='blue',mfc = 'hotpink')
plt.show()
```
