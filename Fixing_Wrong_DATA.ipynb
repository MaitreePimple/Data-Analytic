REPLACING VALUES


> One way to fix this to replace with somthing else


> For small dataset we can delete it one by one but large dataset like below we cannot so we basically put some conditions which makes it easy to edit








```python
import pandas as pd
df = pd.read_csv('data.csv')
print(df.head())
```

```python
# there is no much difference but we change one or 2 value
df.loc[2,'total_rooms'] =9928.0
print(df.head())
```

```python
for x in df.index:
  if df.loc[x, "total_rooms"] > 10000:
    df.loc[x, "total_rooms"] = 10000
    print(df.head())
```

REMOVING ROWS


> Easiest way to do it






```python
for x in df.index:
  if df.loc[x,"total_rooms"] < 10000:
    df.drop(x,inplace = True)
  print(df.head())
```
