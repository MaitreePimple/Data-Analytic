Creating -> bar() fun is used

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,5,6,10])

plt.bar(x,y)
plt.show()
```

```python
x = ["APPLES","BANANAS"]
y =[900,200]
plt.bar(x,y)
plt.show()
```

Horizontal Bars -> barh() function is used

```python
x = np.array(['A','B','C','D'])
y = np.array([3,2,4,5])

plt.barh(x,y)
plt.show()
```

Bar Color

```python
x = np.array(['A','B','C','D'])
y = np.array([3,2,4,5])

plt.bar(x,y,color='red')
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,5,7,9])

plt.bar(x,y, color = "pink")
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,5,7,9])

plt.bar(x,y, color = "#4CAF50")
plt.show()
```

Bar Width -> width argument is used

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

Bar Height -> height argument is used

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```
