plot() -> function is used to draw points(markers) in diagram

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([12,32,45,67])
y = np.array([35,55,23,2])

plt.plot(x,y)
plt.show()
```

Plotting without Line

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([29,82])
y = np.array([4,5])
plt.plot(x,y,'x')
plt.show()
```
