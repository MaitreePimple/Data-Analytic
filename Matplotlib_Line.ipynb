LINESTYLE


> Keyword arg linestyle or short ofrm ls -> to change the style of the plotted line



```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(y, linestyle ='dashed')
```

Shorter Syntax

> linestyle -> ls


> dotted -> :


> dashed -> --







```python
plt.plot(y, ls =':')
```

```python
plt.plot(y,'--')
```

```python
plt.plot(y, ls='-')
```

```python
plt.plot(y,ls='-.')
```

```python
plt.plot(y,ls='or')
```

LINE COLOR


> keyword arg color is used or else c is used -> color of the line



```python
plt.plot(y,ls='--',color='r')
```

```python
plt.plot(y,ls='-.',c='#4CAF50')
```

```python
plt.plot(y,ls='--',c ='hotpink')
```

Line Width

> linewidth / lw -> to change the width of the line



```python
plt.plot(y, c='hotpink',linewidth= '20')
```

Multiple Lines


```python
y1 = np.array([3, 8, 1, 10, 5, 7])
y2 = np.array([10, 20, 30, 40, 50, 60])

plt.plot(y1)
plt.plot(y2)
```

```python
# two lines are added and then plotted over here
x1 = np.array([0,1,2,3])
x2 = np.array([0,1,2,3])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([10, 20, 30, 40])

plt.plot(x1,y1,x2,y2)
```
