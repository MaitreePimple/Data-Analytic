PLOTTING


> plot() -> create diagram


> Pyplot can also be used , its an submodule of the Matplotlib library to visulaize the diagram on the screen






```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
df.plot()
plt.show()
```

SCATTER PLOT


> kind argument is used (kind = "scatter")


> Scatter plot needs an X and Y axies





```python
df = pd.read_csv("data.csv")
print(df.head())
```

```python
df.plot(kind = "scatter", x = "population" , y = "households")
plt.show
```

```python
df.plot(kind = "scatter", x = "longitude" , y = "latitude")
plt.show
```

HISTOGRAM


> kind = "hist"


> histogram needs only one column and it gives us an frequency interval





```python
df['households'].plot(kind = 'hist')
plt.show()
```
