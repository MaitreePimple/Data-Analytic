-> scatter() function is used

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

```python
x = np.array ([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)

x = np.array ([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array ([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)

plt.show()
```

Colors

```python
x = np.array ([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y, color = 'hotpink')

x = np.array ([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array ([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y, color ='#88c999')

plt.show()
```

Color Each Dot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","orange","purple","beige","brown","gray","cyan","magenta","lime"])

plt.scatter(x,y, c=colors)
plt.show()
```

ColorMap -> means list of colours


> How to Use the ColorMap -> cmap argument is used


> There are various type of colours which can be used







```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array ([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y, c=colors, cmap='viridis')
plt.show()
```

```python
x = np.array([4,5,6,7])
y = np.array([11,12,13,14])
colors = np.array([4,67,8,9])
plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

SIZE -> s argument is used

```python
x = np.array([9,8,7,6])
y = np.array([70,80,96,78])
sizes = np.array([20,78,96,56])

plt.scatter(x,y, s=sizes)
plt.show()
```

ALPHA  -> adjust transparency

```python
x = np.array([9,8,7,6])
y = np.array([70,80,96,78])
sizes = np.array([20,78,96,56])

plt.scatter(x,y, s=sizes, alpha=0.5)
plt.show()
```

Combine Color Size and Alpha

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10*np.random.randint(100, size=(100))

plt.scatter(x,y, c= colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```
